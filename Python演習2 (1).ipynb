## 課題1

各問で示されている図を出力してください。

その際、以下の要件を必ず満たすようにしてください。


1.   for文を必ず用いること
2.   変数`size`の値を変えることで自由に図形の大きさを変更できること




### 問1

以下の図形を出力してください。

```
*
**
***
****
*****
```

````python
# 図形の大きさを指定する変数
size = 5

# sizeの数だけ繰り返し処理を行う
for i in range(1, size + 1):
    # 現在の行数分だけアスタリスクを表示する
    print('*' * i)
````

### 問2

以下の図形を出力してください。

```
*
 *
  *
   *
    *
```

````python
# 図形の大きさを指定する変数
size = 5

# sizeの数だけ繰り返し処理を行う
for i in range(size):
    # i個の半角スペースの後にアスタリスクを表示する
    print(' ' * i + '*')
````

### 問3

以下の図形を出力してください。

```
    *
   *
  *
 *
*
```

````python
# 図形の大きさを指定する変数
size = 5

# sizeの数だけ繰り返し処理を行う
for i in range(size):
    # (size - 1 - i)個の空白の後にアスタリスクを表示する
    print(' ' * (size - 1 - i) + '*')
````

### 問4

以下の図形を出力してください。

```
*   *
 * *
  *
 * *
*   *
```

````python
size = 5

# 行(i)のループ
for i in range(size):
    row = ""
    # 列(j)のループ
    for j in range(size):
        # 左上から右下への対角線 (i == j)
        # または、右上から左下への対角線 (i + j == size - 1)
        if i == j or i + j == size - 1:
            row += "*"
        else:
            row += " "
    print(row)
````

### 問5

以下の図形を出力してください。

```
    *
   ***
  *****
 *******
*********
```

````python
size = 5

for i in range(size):
    # 1. 空白の数を計算：下の行に行くほど減る (4, 3, 2, 1, 0)
    spaces = ' ' * (size - 1 - i)
    
    # 2. アスタリスクの数を計算：奇数で増える (1, 3, 5, 7, 9)
    # 式： 2 * i + 1
    stars = '*' * (2 * i + 1)
    
    print(spaces + stars)
````
