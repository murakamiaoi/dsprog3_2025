自然数を入力とし、入力された自然数が3の倍数(Ex. 3, 6, 9, ...)であるときは`hoge`、3を含む数(Ex. 3, 13, 23, ...)であるときは`huga`、両方の条件を満たす時は`piyo`を出力する関数`nabeatsu`を定義してください

```

6

# テストケース2
13

# テストケース3
33
```

````python
def nabeatsu(num: int) -> str:
    # 判定条件の準備
    is_multiple_of_3 = (num % 3 == 0)
    contains_3 = ('3' in str(num))

    # 両方の条件を満たす場合
    if is_multiple_of_3 and contains_3:
        return 'piyo'
    # 3の倍数の場合
    elif is_multiple_of_3:
        return 'hoge'
    # 3を含む数の場合
    elif contains_3:
        return 'huga'
    # いずれにも該当しない場合（任意ですが、通常は数値を返したり空文字を返したりします）
    else:
        return str(num)

case_1 = 6
case_2 = 13
case_3 = 33

print('case_1:', nabeatsu(case_1))
print('case_2:', nabeatsu(case_2))
print('case_3:', nabeatsu(case_3))
````

### 問2

異なる自然数を含む配列とターゲットとなる自然数を入力として、配列に含まれる自然数を足し合わせてターゲットとなる自然数になる組み合わせを出力する関数`combination_sum`を定義してください

このとき、同じ自然数は何度でも利用してよいものとし、もし組み合わせが存在しない場合は空の配列を返してください

例えば、配列`[2, 3, 6, 7]`、ターゲット`7`を入力としたときの出力は`[[2,2,3],[7]]`になります

また、以下のテストケースを入力した際の出力結果を示してください

```
# テストケース1
配列: [2, 3, 6, 7]
ターゲット: 14

# テストケース2
配列: [2, 3, 5]
ターゲット: 8

# テストケース3
配列: [2, 9, 11]
ターゲット: 100
```

````python
def combination_sum(candidates: list[int], target: int) -> list[list[int]]:
    results = []

    def backtrack(remain, combo, start):
        if remain == 0:
            # ターゲットに達したので結果をコピーして保存
            results.append(list(combo))
            return
        elif remain < 0:
            # ターゲットを超えたのでこれ以上探索しない（枝切り）
            return

        for i in range(start, len(candidates)):
            # 同じ数字を何度も選べるよう、次の再帰でも index は i のまま
            combo.append(candidates[i])
            backtrack(remain - candidates[i], combo, i)
            # 探索が終わったら最後の要素を取り除いて別の分岐へ（バックトラック）
            combo.pop()

    # 重複した組み合わせを防ぐため、念のためソートしておくと効率的です
    candidates.sort()
    backtrack(target, [], 0)
    return results

candidates_1 = [2, 3, 6, 7]
target_1 = 14

candidates_2 = [2, 3, 5]
target_2 = 8

candidates_3 = [2, 9, 11]
target_3 = 3 # (テストケース3は100ですが、短い例で動作確認しやすくするため3を例示)
# 実際には 100 を代入して実行します

print('case_1:', combination_sum(candidates_1, target_1))
print('case_2:', combination_sum(candidates_2, target_2))
print('case_3:', len(combination_sum(candidates_3, 100))) # 100は出力が非常に多いため個数のみ確認
````

### 問3

括弧を含む文字列を入力として、括弧の内包関係が正しいかどうかを判定し、正しければ`True`、正しくなければ`False`を返す関数`format_checker`を定義してください

例えば、`[](){}`や`[({})]`は正しい、`[(]`や`{[(})]`は正しくないと判定します

また、以下のテストケースを入力した際の出力結果を示してください

```
# テストケース1
Hello I’m [(firstname) (lastname)] and I’m {age} years old!

# テストケース2
Hello I’m [(firstname) (lastname]) and I’m {age} years old!

# テストケース3
Hello I’m ((firstname) (lastname) and I’m {age} years old!
```

````python
def format_checker(text:str) -> bool:
    return
````

````python
def format_checker(text: str) -> bool:
    # 括弧の対応関係を定義
    bracket_map = {')': '(', ']': '[', '}': '{'}
    open_brackets = bracket_map.values()
    stack = []

    for char in text:
        # 開き括弧ならスタックに追加
        if char in open_brackets:
            stack.append(char)
        # 閉じ括弧ならチェック
        elif char in bracket_map:
            # スタックが空、または対応する開き括弧が一番上にない場合は不正
            if not stack or stack.pop() != bracket_map[char]:
                return False
                
    # 最終的にスタックが空ならTrue（余分な開き括弧がないか確認）
    return len(stack) == 0

case_1 = 'Hello I’m [(firstname) (lastname)] and I’m {age} years old!'
case_2 = 'Hello I’m [(firstname) (lastname]) and I’m {age} years old!'
case_3 = 'Hello I’m ((firstname) (lastname) and I’m {age} years old!'

print('case_1:', format_checker(case_1))
print('case_2:', format_checker(case_2))
print('case_3:', format_checker(case_3))
````
